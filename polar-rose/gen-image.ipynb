```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Set the figure size and DPI for high resolution
plt.figure(figsize=(12, 12), dpi=300)

# Set the background color to black
plt.style.use('dark_background')
plt.gcf().set_facecolor('black')

# Create theta values (more points = smoother curve)
theta = np.linspace(0, 2*np.pi, 1000)

# Number of petals
n = 7

# Number of layers for gradient effect
num_layers = 15

# Create custom color gradients (you can modify these colors)
# Format: [(position, (R,G,B,alpha)), ...]
colors_main = [
    (0, (0, 0.5, 1, 1)),    # Blue
    (1, (0.5, 0, 1, 0))     # Purple
]
custom_cmap = LinearSegmentedColormap.from_list('custom', colors_main)

# Plot multiple layers with decreasing size and varying colors
for i in range(num_layers):
    # Calculate size factor for this layer
    size_factor = 1 - (i * 0.02)
    
    # Calculate r for this layer
    r = size_factor * np.cos(n*theta)
    
    # Convert to Cartesian coordinates
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    
    # Get color from colormap
    color = custom_cmap(i/num_layers)
    
    # Plot this layer
    plt.plot(x, y, color=color, linewidth=2.5, alpha=0.7)

# Remove axes for cleaner look
plt.axis('off')

# Equal aspect ratio ensures the rose looks properly proportioned
plt.axis('equal')

# Tight layout to remove excess margins
plt.tight_layout()

# Save with black background
plt.savefig('polar_rose_gradient.png', 
            bbox_inches='tight', 
            pad_inches=0, 
            facecolor='black', 
            edgecolor='none')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the figure size and DPI for high resolution
plt.figure(figsize=(12, 12), dpi=300)

# Set the background color to black
plt.style.use('dark_background')
plt.gcf().set_facecolor('black')

# Create time values
t = np.linspace(0, 12*np.pi, 1000)

# Create complex exponential
z = np.exp(1j * t)

# Calculate the spiral effect
spiral = z * t/10

# Separate real and imaginary parts
x = spiral.real
y = spiral.imag

# Create gradient colors
colors = np.linspace(0, 1, len(t))

# Plot with gradient effect
plt.scatter(x, y, c=colors, 
           cmap='plasma',  # Try different colormaps like 'viridis', 'magma', 'cividis'
           s=1.5,         # Size of points
           alpha=0.6)     # Transparency

# Remove axes for cleaner look
plt.axis('off')

# Equal aspect ratio
plt.axis('equal')

# Tight layout
plt.tight_layout()

# Save with black background
plt.savefig('complex_exponential.png', 
            bbox_inches='tight', 
            pad_inches=0, 
            facecolor='black', 
            edgecolor='none')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the figure size and DPI for high resolution
plt.figure(figsize=(12, 12), dpi=300)

# Set the background color to black
plt.style.use('dark_background')
plt.gcf().set_facecolor('black')

# Number of spiral arms
n_arms = 6

# Create time values
t = np.linspace(0, 8*np.pi, 1000)

# Create multiple spiral arms
for i in range(n_arms):
    # Phase shift for each arm
    phase = 2 * np.pi * i / n_arms
    
    # Create complex exponential with phase shift
    z = np.exp(1j * (t + phase))
    
    # Calculate the spiral effect
    spiral = z * t/12
    
    # Separate real and imaginary parts
    x = spiral.real
    y = spiral.imag
    
    # Plot with color gradient
    plt.plot(x, y, 
            alpha=0.6,
            linewidth=1.5,
            color=plt.cm.viridis(i/n_arms))  # Use color gradient

# Remove axes for cleaner look
plt.axis('off')

# Equal aspect ratio
plt.axis('equal')

# Tight layout
plt.tight_layout()

# Save with black background
plt.savefig('complex_exponential_multi.png', 
            bbox_inches='tight', 
            pad_inches=0, 
            facecolor='black', 
            edgecolor='none')

# Display the plot
plt.show()
```
